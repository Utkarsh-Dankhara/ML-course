# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10))   #(width, height)

# 3. Plot data
ax.plot(x, y)

# 4.Customize plot
ax.set(title = "Simple Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & Show
fig.savefig("sample-plot.png")
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                              ncols = 2,
                                              figsize = (10, 5))

# Plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_prices.keys(), nut_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# Plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_prices.keys(), nut_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Now we use Object Oriented matplotlib

## What to use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use pyplot method
* When plotting something advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Pyplot method, AKA plotting directly from the dataframe
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            figsize = (10, 6));
# c is for colour, we will colour the graph on the bases of values in target column
```

```python
# OO method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y = 'chol',
            c = 'target',
            ax = ax);

#ax.set_xlim([45, 100]);
```

```python
# The real OO method from scratch, the previous one was kinda a fake
fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target']);

# Customize the plot
ax.set(title = 'Heart disease and cholestrol levels.',
      xlabel='Age',
      ylabel='Cholestrol');

# Add a legend
ax.legend(*scatter.legend_elements(), title = "target")
# * says, grab the values from scatter, unpack all the values of the target column
# It gets the legend from the 'c' parameter

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
over_50.head()
```

```python
# Subplot of chol, thalach, age
fig, (ax0, ax1) = plt.subplots(figsize = (10, 10),
                              nrows = 2,
                              ncols = 1,
                              sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

#Customize ax0
ax0.set(title = 'Heart disease and Cholestrol levels',
# We removed the xlabel as we have sharex above
       ylabel='Cholestrol')

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a mean line
ax0.axhline(y = over_50['chol'].mean(),linestyle = '--')


#Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target']);

#Customize ax1
ax1.set(title='Heart disase and thalach levels',
       xlabel='Age',
       ylabel='Thalach',)

# Add a legend line
ax1.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line
ax1.axhline(y = over_50['thalach'].mean(),
           linestyle = '--');

# Add a title for the whole figure
fig.suptitle("Heart disease analysis", fontsize = 16, fontweight = "bold");
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize = (10, 6))

# Plot the data
scatter = ax.scatter(x = over_50['age'],
                    y = over_50['chol'],
                    c = over_50['target'],
                     cmap = 'plasma');    # Cmap is colour map to change the colour scheme

# Customize the plot
ax.set(title = 'Heart disease and cholestrol levels.',
      xlabel='Age',
      ylabel='Cholestrol');

# Add a legend
ax.legend(*scatter.legend_elements(), title = "target")
# * says, grab the values from scatter, unpack all the values of the target column
# It gets the legend from the 'c' parameter

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle = '--');
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, thalach, age
fig, (ax0, ax1) = plt.subplots(figsize = (10, 10),
                              nrows = 2,
                              ncols = 1,
                              sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap = 'plasma');

# Change the x axis limit
ax0.set_xlim(50, 80)

#Customize ax0
ax0.set(title = 'Heart disease and Cholestrol levels',
# We removed the xlabel as we have sharex above
       ylabel='Cholestrol')

# Add legend to ax0
ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a mean line
ax0.axhline(y = over_50['chol'].mean(),linestyle = '--')


#Add data to ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap = 'winter');

#Customize ax1
ax1.set(title='Heart disase and thalach levels',
       xlabel='Age',
       ylabel='Thalach',)

# Change ax1 y axis limit
ax1.set_ylim(60, 200)

# Add a legend line
ax1.legend(*scatter.legend_elements(), title = 'Target')

# Add a mean line
ax1.axhline(y = over_50['thalach'].mean(),
           linestyle = '--');

# Add a title for the whole figure
fig.suptitle("Heart disease analysis", fontsize = 16, fontweight = "bold");
```

## Creating Data

```python
# Column from series
seats = pd.Series([5, 5, 5, 5, 5])
# Names like seats_columns don't work as they seem to be non defined even after storing a series to it
```

```python
# New column named seats
sales["Seat"] = seats
sales
#We can also do, sales["Seat"] = pd.Series([5, 5, 5, 5, 5])
```

```python
sales.fillna({"Seat" : 5}, inplace = True)
sales
```

```python
# Another way to make a column is from a python list
fuel_economy = (7.8, 8.6, 6.3, 7.4, 9.1, 5.7, 7.2, 6.8, 4.2, 8.8)
sales["Fuel per 100KM"] = fuel_economy
sales
```

```python
sales["Total Fuel used"] = sales["Odometer (KM)"] / 100 * sales["Fuel per 100KM"]
sales["Total Fuel used (L)"] = sales["Odometer (KM)"] / 100 * sales["Fuel per 100KM"]
sales
```

```python
sales["Wheels"] = 4
sales
```

```python
sales["Passed Road Safety"] = True
sales
```

```python
sales.drop("Total Fuel used", axis = 1, inplace = True)
```
